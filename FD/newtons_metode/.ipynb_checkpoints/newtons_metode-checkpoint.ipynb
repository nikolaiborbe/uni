# Implementering av Newtons metode

Vi skal i dette notatet sjå korleis vi kan implementere Newtons metode i ein og fleire variablar, og korleis vi kan løyse likningssystem ved hjelp av metoden.

## Einvariabel-tilfellet

Vi repeterer fort Newtons metode i ein variabel. Vi startar med å gjette på ei løysing $x_0$ på likninga $$f(x)=0.$$ Så oppdaterer vi gjettinga vår via formelen $$x_{n+1}=x_n-\frac{f(x_n)}{f'(x_n)}.$$ Vi gjer denne prosessen igjen til den konvergerer. Med eit par linjer med kode kan vi implementere Newtons metode som ein funksjon som tek inn $x_0$, $f$, $f'$ og $\varepsilon$, der $\varepsilon$ er nøyaktigheita vi ønsker oss, og gjev oss eit nullpunkt til $f$. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Kode: Newtons metode i ein variabel

```python
def newton(x_0,f,Df,epsilon): #Første gjett, funksjonen, den deriverte, nøyaktigheit
    differanse=np.abs(f(x_0))
    while differanse>epsilon:
        x=x_0-f(x_0)/Df(x_0) #Newtons metode
        differanse=np.abs(f(x)) #Sjekk om vi har konvergert
        x_0=x #Oppdater x_0 til neste iterasjon
    return x
```

### Døme:

Vi ønsker å finne tjuenderota til eit stort tal, la oss seie $99997613$. Det er såklart ingen måte vi kan gjere dette for hand på, talet er alt for stort til at vi kan rekne med det. Vi ønsker derfor å finne tjuenderota numerisk, med Newtons metode. Vi kan løyse dette problemet i to steg:

$\textbf{Steg 1: Skriv om problemet til eit nullpunktsproblem.}$ Dette er ikkje så vanskeleg: Vi vil finne ein $x$ slik at $x^{20}=99997613$, med andre ord, vi vil finne ein $x$ slik at $x^{20}-99997613=0$. Altså finn vi nullpunktet til funksjonen $f(x)=x^{20}-99997613$. 

$\textbf{Steg 2: Gjer Newtons metode.}$ Funksjonen vi har implementert tek inn fire variablar. Vi veit frå steg 1 kva funksjon vi må setje inn, og dermed også kva deriert vi skal setje inn. Vi startar på $x_0=1$, sidan vi veit at tjuenderota må vera større enn $1$, men ikkje veldig mykje større. Vi vel presisjonen $\epsilon=0.0001$. No køyrer vi koden med desse verdiane.

```python
def f(x):
    return x**20-99997613
def Df(x):
    return 20*x**19
x_0=1
epsilon=0.0001

x=newton(x_0,f,Df,epsilon)
print(x)
```

Dei 20 første desimalane i $\sqrt[20]{99997613}$ er $2.51188343353913222397$, så vi ser at sjølv om vi berre har presisjon på $0.0001$ treff vi ganske bra på nullpunktet.

## Utviding til fleire variablar

La oss no sjå på det fleirdimensjonale tilfellet. Vi vil finne nullpunkta til eit vektorfelt $$F:\mathbb{R}^n\rightarrow\mathbb{R}^n.$$ Vi startar med å gjette på eit nullpunkt $\mathbf{x_0}$, og oppdaterer gjettinga vår via uttrykket $$\mathbf{x_{n+1}}=\mathbf{x_n}-\left(F'(\mathbf{x_n})\right)^{-1}\cdot F(\mathbf{x_n}).$$ Koden for den fleirvariable metoden er ganske lik den einvariable koden, men vi bruker nokon spesialfunksjonar for matrisemultiplikasjon og -invers, og lengde av ein vektor. Koden ser de under. Merk at denne koden fungerer like fint i to og tre variablar, og i $200$ variablar, for den saks skuld.

#### Kode: Newtons metode i fleire variablar

```python
def FlVNewton(x_0: list, F, DF, epsilon: float) -> list:
    differanse=np.linalg.norm(F(x_0))
    while differanse>epsilon:
        DFI=np.linalg.inv(DF(x_0))#Hjelpesteg
        x=x_0-np.matmul(DFI,F(x_0)) #Newtons metode
        differanse=np.linalg.norm(F(x)) #Sjekk om vi har konvergert
        x_0=x #Oppdater x_0 til neste iterasjon
    return x
    
```

$\textbf{Oppgåve:}$ Bruk Newtons metode til å finne eit nullpunkt til funksjonen under. Bruk det spesifiserte startpunktet. Du kan velge nøyaktigheit sjølv. Du kan også prøve å velje eit anna startpunkt, og sjå kva du får.

Etter som den oppgitte funksjonen,

$F(x, y) = (x^2+y+1, e^{x+y})$

ikke har røtter, velger jeg å bruke funksjonen som tilsynelatende er feil, men som er skrevet i koden,

$F(x,y)=(x^2 \cdot y +1, e^{x}+y)$

```python
#Dette er funksjonen F(x,y)=(x^2+y+1,e^(x+y)).
def F(x: list) -> list:
    return (x[0]**2*x[1]+1,np.exp(x[0])+x[1])

def DF(x: list) -> np.array:
    return np.array([[2*x[0]*x[1], x[0]**2], [np.exp(x[0]), 1]], np.float64)
    
x_0=[-0.7,-2.5]
epsilon = 0.0001

x = FlVNewton(x_0, F, DF, epsilon)
x
```

$\textbf{Oppgåve:}$ Finn minst to løysinger på likningssystemet $$\begin{cases} x^2+y^2+z^2=9 \\ x^2-y^2+2z^2=1 \\ x+y+10z=1\end{cases} $$ ved hjelp av Newtons metode.

```python
def F(xl: list) -> list:
    x = xl[0]; y = xl[1]; z = xl[2]
    F_1 = x**2 + y**2 + z**2 - 9
    F_2 = x**2 - y**2 + 2*z**2 - 1
    F_3 = x + y + 10*z - 1
    return (F_1, F_2, F_3)

def DF(xl: list) -> np.array:
    x = xl[0]; y = xl[1]; z = xl[2]
    out = np.array([
        [2*x, 2*y, 2*z],
        [2*x, -2*y, 4*z],
        [1, 1, 10]
    ], np.float64)
    return out

epsilon = 0.0001

x_0 = [1., 1., 1.]
x_1 = FlVNewton(x_0, F, DF, epsilon)

x_0 = [-1, -1, -1]
x_2 = FlVNewton(x_0, F, DF, epsilon)
x_1, x_2
```

### Numerisk analyse

Dersom vi har lyst til å gjera meir analyse enn berre å finna nullpunktet kan vi modifisera funksjonen til å gje oss meir informasjon. Under har vi definert ein funksjon som ikkje berre finn nullpunktet til funksjonen, men som også gjev oss plotter følgjene $\{x_n\}$ og $\{f(x_n)\}$ som funksjonar av $n$.

```python
def PlotNewton(x_0,f,Df,epsilon): #Første gjett, funksjonen, den deriverte, nøyaktigheit
    differanse=np.abs(f(x_0))
    n=[]
    xs=[]
    d=[]
    i=0
    while differanse>epsilon:
        x=x_0-f(x_0)/Df(x_0) #Newtons metode
        differanse=np.abs(f(x)) #Sjekk om vi har konvergert
        #Legg til element i listene vi skal plotte
        n.append(i)
        xs.append(x)
        d.append(differanse)
        #Oppdater variablane til slutt
        x_0=x #Oppdater x_0 til neste iterasjon
        i+=1
    fig, ax = plt.subplots(1,2,figsize=(9,3))
    ax[0].semilogy(n,xs) #Bruker semilog fordi tala er store
    #ax[0].plot(n,xs) #Bruk denne dersom tala ikkje er så store
    ax[1].semilogy(n,d) #Bruker semilog fordi tala er store
    #ax[1].semilogy(n,d) #Bruk denne dersom tala ikkje er så store
    ax[0].set_xlabel('Iterasjonar')
    ax[0].set_ylabel('Verdien av x_n')
    ax[1].set_xlabel('Iterasjonar')
    ax[1].set_ylabel('Verdien av f(x_n)')
    plt.show()
    print("Iterasjoner brukt: ", i)
    print(x)
    #print(xs) #Å printe ut lista med gjett er valfritt, men kan vera nyttig dersom vi skal analysere val
    #av startgjett og presisjon
    #print(d) # Same som for den andre lista
```

$\textbf{Oppgåve:}$ Juster koden slik at funksjonen også printer ut antal iterasjonar den brukte på å konvergere.

$\textbf{Utfordring:}$ Lag ein tilsvarande funksjon for den fleirvariable versjonen av Newtons metode.

```python
x = PlotNewton(1, f, Df, 0.0001)
```

## Uventa fenomen i Newtons metode

Newtons metode finn nullpunkta til ein funksjon, men berre dersom krava for Kantorovitsj-teoremet er oppfylt. Dersom eit av krava ikkje er oppfylt kan uventa ting skje. Vi fokuserer på døme i ein variabel, men dei same fenomena dukker også opp i fleire variablar.

### Døme 1: Startpunktet er for langt unna eit nullpunkt.

Vi vil finne nullpunktet til $\arctan(x)$ med Newtons metode. Vi veit at nullpunktet er i $x=0$, så vi venter at Newtons metode skal konvergere mot dette punktet. La oss gjere Newtons metode med startpunkt $x_0=2$. $$x_1=2-\arctan(2)(1+2^2)=2-5\arctan(2)\approx -3.54,$$ så $x_1$ er faktisk lenger unna nullpunktet enn $x_0$. Dette er ikkje nødvendigvis eit problem, men dersom vi ser på eit generelt ledd har vi at $$-x_{n+1}=\arctan(x_n)(1+x_n^2)-x_n\geq 2\arctan(x_n)x_n-x_n=x_n(2\arctan(x_n)-1),$$ og sidan $(2\arctan(x_n)-1)>1$ for alle $|x_n|\geq 2$ konkluderer vi med at $$|x_{n+1}|\geq |x_n|,$$ og at Newtons metode dermed divergerer for $x_0=2$. De kan sjølv sjekke at for $x_0=1$ vil Newtons metode konvergere mot $0$, så problemet her er at punktet vi har valgt er for langt unna nullpunktet til at metoden konvergerer. 

### Døme 2: Funksjonen er ikkje deriverbar i nullpunktet

Ein annan måte Newtons metode kan svikte oss på er dersom funksjonen vår ikkje er fin nok. Dersom nullpunktet vi leiter etter også er eit kritisk punkt(altså at $F'(\mathbf{0})$ ikkje er inverterbar) kan vi få problem. Vi illustrerer dette med eit døme i ein variabel. Sjå på funksjonen $$f(x)=\begin{cases} \sqrt{x},\quad x\geq 0, \\ -\sqrt{-x},\quad x<0.\end{cases}$$ Merk at denne funksjonen har nullpunktet sitt i $x=0$, men merk også at $f$ ikkje er deriverbar i $0$, sidan grenseverdien som definerer den deriverte er ulik frå høgre og venstre(gjerne plott funksjonen $f$ dersom du ikkje er overtydd om dette). Ellers har vi at $$f'(x)=\begin{cases} \frac{1}{2\sqrt{x}},\quad x>0,\\ -\frac{1}{2\sqrt{-x}},\quad x<0.\end{cases}$$

La oss no bruke Newtons metode på $f$. Dersom vi vel eit startpunkt $x_0>0$ får vi at $$x_1=x_0-\frac{f(x_0)}{f'(x_0)}=x_0-\frac{\sqrt{x_0}}{1/\left(2\sqrt{x_0}\right)}=-x_0,$$ og vidare får vi $$x_2=x_1-\frac{f(x_1)}{f'(x_1)}=-x_0-\frac{f(-x_0)}{f'(-x_0)}=-x_0-\frac{-\sqrt{-x_0}}{-1/\left(2\sqrt{-x_0}\right)}=x_0.$$
Her kjem vi altså berre til å gå fram og tilbake mellom $x_0$ og $-x_0$. Det same vil skje dersom vi vel $x_0<0$.

### Døme 3: To nærliggjande punkt kan konvergere mot ulike nullpunkt

Til slutt skal vi sjå på eit fenomen som er ganske merkeleg, nemleg at sjølv om Newtons metode konvergerer mot eit nullpunkt, er det ikkje nødvendigvis det nullpunktet vi såg for oss. Som eit døme nytter vi Newtons metode på funksjonen $$F(x,y)=\begin{pmatrix} y-x^3-2x^2+1 \\ y+x^2-1 \end{pmatrix}.$$ Denne funksjonen har tre nullpunkt, som er markerte med ei stjerne i plottet under. Vidare er det markert fleire punkt i planet, i tre ulike fargar. Fargane seier noko om kva nullpunkt Newtons metode konvergerer til dersom vi startar i dette punktet.
