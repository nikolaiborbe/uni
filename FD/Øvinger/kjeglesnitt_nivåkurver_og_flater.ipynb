# Kjeglesnitt, nivåkurver og flater

I dette notatet skal vi ta for oss plotting av kjeglesnitt, nivåkurver og flater i 3D.

```python
# Importer pakkene vi trenger
import matplotlib.pyplot as plt
import numpy as np
```

### Oppvarming: funksjoner av én variabel

Som oppvarming plotter vi et par funksjoner av én variabel. Da kan vi bruke funksjonen plot(x,y), hvor $y$ er gitt som en funksjon av $x$.

```python
# Plotting av en sinusfunksjon
#plt.style.use('_mpl-gallery')

# Genererer x-aksen og definerer y som en funksjon av x
x = np.linspace(0, 10, 100)
y = 4 + 2 * np.sin(2 * x)

# Plot
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
# Plotting av f(x)=(x-2)^2+1

# Genererer x-aksen
x = np.linspace(-5, 5, 100)

# Plot
fig, ax = plt.subplots()

ax.plot(x, (x-2)**2+1, linewidth=2.0) # Her har vi definert funksjonen y inne i plot-funksjonen

ax.set(xlim=(-2, 5), xticks=np.arange(-2, 5),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

## Kjeglesnitt

Et generelt kjeglesnitt kan ikke beskrives som grafen til en funksjon, så da nytter det ikke å bruke plot(x, y) slik som ovenfor. Dersom vi vil plotte slike kjeglesnitt kan vi benytte oss av contour-funksjonen i stedet.

```python
# Plotting av x^2+y^2=4
#plt.style.use('_mpl-gallery-nogrid') # Her plotter vi uten rutenett, men det er valgfritt

# Genererer aksene og kjeglesnittet
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = X**2+Y**2-4 # Flytt hele kjeglesnittet over til den ene siden av ligningen
levels=[0] # Dette er fordi vi vil vise kjeglesnittet 0=Z=X^2+Y^2-4

# Plot
fig, ax = plt.subplots()
ax.contour(X, Y, Z, levels=levels, colors='blue')
ax.set_aspect('equal') # Justerer forholdet mellom x- og y-aksen. 

plt.show()
```

```python
# Plotting av 4x^2-2y^2=3

# Genererer aksene og kjeglesnittet
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = 4*X**2-2*Y**2-3 
levels=[0] 

# Plot
fig, ax = plt.subplots()
ax.contour(X, Y, Z, levels=levels, colors='blue')
ax.set_aspect('equal')

plt.show()
```

```python
# Plotting av x-4y^2=2

# Genererer aksene og kjeglesnittet
X, Y = np.meshgrid(np.linspace(0, 6, 256), np.linspace(-3, 3, 256)) # Justerer litt på området vi plotter
Z = X-4*Y**2-2 
levels=[0] 

# Plot
fig, ax = plt.subplots()
ax.contour(X, Y, Z, levels=levels, colors='blue')
ax.set_aspect('equal')

plt.show()
```

## Nivåkurver

Dersom du har studert koden vi har brukt til plotting av kjeglesnitt, har du kanskje lagt merke til at vi egentlig plotter nivåkurven til en skalar-funksjon på nivå $0$. Vi kan bruke samme kode til å plotte mer generelle nivåkurver.

```python
# Nivåkurver til sin(xy)

# Genererer aksene og funksjonen
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = np.sin(X*Y)
levels=0 # Vi ser altså på nivåkurven f(x,y)=0

# Plot
fig, ax = plt.subplots()
ax.contour(X, Y, Z, levels=levels, colors='blue')
ax.set_aspect('equal')

plt.show()
```

Vi kan også plotte flere nivåkurver i samme figur.

```python
# Genererer aksene of funksjonen
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = np.sin(X*Y)

# Nivåene
levels=[0,.5] # Nivåene må være i én vektor, og i stigende rekkefølge

# Plot
fig, ax = plt.subplots()
ax.contour(X, Y, Z, levels=levels,colors=['blue','red']) # Vi spesifiserer fargen til hvert nivå i colors-vektoren
# Dersom du ikke har med colors-vektoren vil du få standardfarger
ax.set_aspect('equal')

plt.show()
```

## Flater i 3D

Dersom flaten vi vil plotte er en graf, altså at en av variablene er en funksjon av de to andre, kan vi bruke plot_surface.

```python
# Plotting av flaten z=sin(sqrt(x^2+y^2))
from matplotlib import cm # For å kunne velge farge på flaten

# Genererer aksene og funksjonen
X = np.arange(-5, 5, 0.24)
Y = np.arange(-5, 5, 0.24)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)/abs(X**2 + Y**2)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)


plt.show()
```

```python
# Plotting av planet x+2y-3z=-3 

# Genererer aksene og funksjonen
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z=X/3+2*Y/3+1 # Merk: Dette er ligningen til planet, vi har bare løst ligninga med hensyn på z

# Plot 
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)

plt.show()
```

```python
# Plotting av kjeglen z=sqrt(x^2+y^2)

# Genererer aksene og funksjonen
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.sqrt(X**2+Y**2)

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)

plt.show()
```

Dette ser ikke helt ut som en kjegle. Grunnen til dette er at vi har plottet kjeglen på et firkantet domene, i stedet for et rundt domene, som ville vært mer naturlig. Vi kan bruke parametrisering av domenet til å plotte flater på mer naturlige domener. For kjeglen er det naturlig å bruke polarkoordinater.

```python
# Lager et rutenett i polarkoordinater og definerer funksjonen 
r = np.linspace(0, 5, 50)
p = np.linspace(0, 2*np.pi, 50)
R, P = np.meshgrid(r, p)
Z = R

# Uttrykker xy-planet med polarkoordinater
X, Y = R*np.cos(P), R*np.sin(P)

# Plotter
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.Blues)

plt.show()
```

### En flate som ikke er en graf

For andregradsflater som ikke er grafer, for eksempel kuler og ellipsoider, må vi bruke andre metoder. Ett alternativ er å bruke kulekoordinater (også kalt sfæriske koordinater). Dette står godt forklart i avsnitt 8.11 i boka, og vi kommer til å gå gjennom det senere i kurset. 

```python
# Kulekoordinater
U= np.arange(0, 2*np.pi+0.1, 0.1) # Lengdegrad
V =np.arange(0, np.pi+0.05, 0.05) # Breddegrad

# Kartesiske koordinater uttrykt ved kulekoordinater
X=2*np.outer(np.cos(U),np.sin(V)) 
Y=2*np.outer(np.sin(U), np.sin(V))
Z=2*np.outer(np.ones(np.size(U)), np.cos(V))

#Plot
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)

plt.show()
```

```python
x = np.linspace(-10, 10, 40)
y = np.linspace(-10, 10, 40)
X, Y = np.meshgrid(x, y)
Z = np.where(X != Y, (X**3 - Y**3)/X**2 + Y**2, 0)

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(X, Y, Z, cmap="viridis", edgecolor="none")
plt.show()
```
