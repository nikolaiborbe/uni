```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
from fractions import Fraction
```

```python
#parametere
m = 9.11e-31
e = 1.602e-19
```

```python
def r(B, U_a):
    return (1/B) * sqrt((2*U_a)/(e/m))
```

### Oppgave 5
Sett opp i et kurvediagram kurveskarer over r sfa. $\vec B$ i området $0 < B < 15$ gauss for
$U_a = 20V, 40V, 60V$. Bruk for eksempel Python eller Matlab

```python
B = np.arange(0, 15, 0.25)
plt.plot(B, r(B, 20), label="$U_a = 20$")
plt.plot(B, r(B, 40), label="$U_a = 40$")
plt.plot(B, r(B, 60), label="$U_a = 60$")
plt.ylabel("$r [m]$")
plt.xlabel("$\\vec B [G]$")
plt.legend()
plt.show()
```

### Oppgave 6
Bruk kurveskaren som du satt opp i oppgaven ovenfor til å anslå gunstig magnetfelt $\vec B$ for
tre ulike akselerasjonsspenninger $U_a = 20V, 40V, 60V$.

```python
B = np.linspace(0.0004, 0.001, 50)
plt.plot(B, r(B, 20), label="$U_a = 20$")
plt.plot(B, r(B, 40), label="$U_a = 40$")
plt.plot(B, r(B, 60), label="$U_a = 60$")
plt.ylabel("$r [m]$")
plt.xlabel("$\\vec B [G]$")
plt.legend()
plt.show()
```

Det virker som om at bra utgangspunkt for $\vec B$ kan være å sette $\vec B = [0.0003G, 0.001G]$.
