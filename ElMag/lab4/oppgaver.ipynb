```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
```

```python
#parametere
m = 9.11e-31
e = 1.602e-19
```

```python
def r(B, U_a):
    return (1/B) * sqrt((2*U_a)/(e/m))
```

### Oppgave 5
Sett opp i et kurvediagram kurveskarer over r sfa. $\vec B$ i omr√•det $0 < B < 15$ gauss for
$U_a = 20V, 40V, 60V$. Bruk for eksempel Python eller Matlab

```python
B = np.arange(0, 15, 0.25)
plt.plot(B, r(B, 20)*1e4, label="$U_a = 20$")
plt.plot(B, r(B, 40)*1e4, label="$U_a = 40$")
plt.plot(B, r(B, 60)*1e4, label="$U_a = 60$")
plt.ylabel("$r [m]$")
plt.xlabel("$\\vec B [G]$")
plt.legend()
plt.show()
```
