I denne oppgaven skal du skrive to funksjoner. Den f칮rste skal ta inn to argumenter `x` og `y`, lage en variabel `num` = `x//y`, og ***returnere*** resultatet slik at programmet kan skrive til skjerm "Heltallsdivisjon av `x` over `y` gir `num`".

Den andre funksjonen skal ta inn ett argument x, lage en varibel num = x\*\*2, og ***returnere*** resultatet slik at programmet kan skrive til skjerm "Kvadratet av `x` er `num`".

***Skriv koden din i kodeblokken under***

````python
def en(x, y):
    num = x // y
    return x, y, num

def to(x):
    num = x**2
    return x, num

print("Heltallsdivisjon av %d over %d gir %d" % en(4, 2))
print("Kvadratet av %d er %d" % to(4))
````
